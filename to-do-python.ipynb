# 1. visualizar tarefas
# 2. adicionar tarefas
# 3. atualizar tarefas   # escolher o id da tarefa, depois descricao, data, categoria
# 4. remover tarefas
# 0. fechar

```python
def show_menu():
    print(f'''
    Bem vindo ao PyToDo!\n

    Digite a opção desejada:
    1. Visualizar tarefas
    2. Adicionar tarefas
    3. Atualizar tarefas
    4. Remover tarefas
    0. Fechar
    ''')
    new_option = -1
    while new_option != "1" and new_option != "2" and new_option != "3" and new_option != "4" and new_option != "0":
        new_option = input('Digite a opção desejada: ')

    return int(new_option)

def check_to_do():
    print("--- 1.VISUALIZAR TO DOS ---")
    for i in tarefas.keys():
        print(f'''
        {i} | {tarefas[i][0]} | {tarefas[i][1]} | {tarefas[i][2]}
        ''')

def add_to_do():
    print("--- 2.ADICIONAR TO DOS ---")
    new_id =  input('Digite o ID da nova tarefa: ')

    while tarefas.get(new_id) != None:
        new_id =  input('ID JÁ EXISTENTE! Digite o ID da nova tarefa: ')
    
    lista = []

    lista.append(input('Digite a descrição da nova tarefa: '))
    lista.append(input('Digite a data da nova tarefa: '))
    lista.append(input('Digite a categoria da nova tarefa: '))

    tarefas[new_id] = lista
    print("TAREFA ADICIONADA COM SUCESSO!")

def update_to_do():
    print("--- 3.ATUALIZAR TO DOS ---")
    update_id =  input('Digite o ID da tarefa que deseja atualizar: ')

    while tarefas.get(update_id) == None:
        update_id =  input('ID NÃO EXISTENTE! Digite o ID da tarefa a atualizar: ')
    
    lista = []

    lista.append(input('Digite a nova descrição da tarefa: '))
    lista.append(input('Digite a nova data da tarefa: '))
    lista.append(input('Digite a nova categoria da tarefa: '))

    tarefas[update_id] = lista
    print("TAREFA ATUALIZADA COM SUCESSO!")

def remove_to_do():
    print("--- 4.REMOVER TO DOS ---")
    remove_id =  input('Digite o ID da tarefa que deseja remover: ')

    while tarefas.get(remove_id) == None:
         remove_id =  input('ID NÃO EXISTENTE! Digite o ID da tarefa a remover: ')
    
    del tarefas[remove_id]
    print("TAREFA REMOVIDA COM SUCESSO!")



option = -1
tarefas = {"ID": ["DESCRIÇÃO", "DATA", "CATEGORIA"]}

while option != 0:

    if option == 1:
        check_to_do()
    elif option == 2:
        add_to_do()
    elif option == 3:
        update_to_do()
    elif option == 4:
        remove_to_do()
    
    option = show_menu()

print("----- PROGRAMA ENCERRADO -----")
```
