Parte 1 - Sistema de Controle

Nesta parte do projeto você deverá construir um sistema para realizar o controle da vacinação. Você deverá considerar que existem somente os seguintes tipos de vacina: Pfizer, CoronaVac, Astrazeneca.

O programa deverá exibir um menu inicial com as seguintes opções:

Registrar uma vacinação.
Nesta opção você deverá perguntar:
O nome da pessoa.
Qual a vacina.
Você deverá salvar além desses dados, o valor do dia e da hora de vacinação.
Adicionar estoque de uma vacina.
Nesta opção você deverá perguntar:
O nome da vacina.
A quantidade a ser adicionada.
Obter número total de vacinados.
Nesta opção você deverá imprimir a quantidade total de vacinados e além disso a quantidade total de vacinados por cada uma das vacinas.
Obter média de vacinação diária.
Nesta opção você deverá imprimir a média diária somando todas as vacinas e além disso a média diária por cada uma das vacinas.
Obter a quantidade atual de doses de um tipo de vacina.
Você deverá imprimir a quantidade restante de doses para cada uma das vacinas.
Fechar programa.
Observações:
Caso a pessoa digite uma opção inválida, a programa deverá imprimir "Opção inválida" e perguntar novamente uma nova opção.

```python
def imprime_menu():
    print(f'''
    Bem vindo ao PythonVacina!\n

    Digite a opção desejada:
    [1] Registrar vacinação
    [2] Incluir estoque de vacina
    [3] Total de vacinados
    [4] Média de vacinação diária
    [5] Quantidade atual de vacina por marca
    [6] Fechar
    ''')

def registrar_vacinacao():
    nome = input('Digite o nome: ')
    vacina = int(input(f'''Digite a vacina escolhida:
[1] Pfizer, estoque: {estoque_por_vacina[0]}
[2] Coronavac, estoque: {estoque_por_vacina[1]}
[3] Astrazeneca, estoque: {estoque_por_vacina[2]}
[4] Encerrar'''))

    while vacina != 1 and vacina != 2 and vacina != 3 and vacina != 4:
        print('Opção Inválida!')
        vacina = int(input(f'''Digite a vacina escolhida:
[1] Pfizer, estoque: {estoque_por_vacina[0]}
[2] Coronavac, estoque: {estoque_por_vacina[1]}
[3] Astrazeneca, estoque: {estoque_por_vacina[2]}
[4] Encerrar'''))

    if vacina == 4:
        return
    
    for i in range(0,3):
        if vacina == i + 1:
            if estoque_por_vacina[i] > 0:
                data = input("Digite a data: ")
                hora = input("Digite a hora: ")

                vacinados[nome] = [data, hora]
                vacinados_por_vacina[i] += 1
                estoque_por_vacina[i] -= 1
            else:
                print('Não há estoque da vacina escolhida. Tente novamente!')

def incluir_estoque():
    vacina = int(input(f'''Digite a vacina para incluir estoque:
[1] Pfizer, estoque: {estoque_por_vacina[0]}
[2] Coronavac, estoque: {estoque_por_vacina[1]}
[3] Astrazeneca, estoque: {estoque_por_vacina[2]}
[0] Encerrar'''))

    while vacina != 1 and vacina != 2 and vacina != 3 and vacina != 4:
        print('Opção Inválida!')
        vacina = int(input(f'''Digite a vacina para incluir estoque:
[1] Pfizer, estoque: {estoque_por_vacina[0]}
[2] Coronavac, estoque: {estoque_por_vacina[1]}
[3] Astrazeneca, estoque: {estoque_por_vacina[2]}
[4] Encerrar'''))

    if vacina == 4:
        return
    
    for i in range(0,3):
        if vacina == i + 1:
            qtd = int(input(f'Digite a quantidade de doses a serem disponibilizadas de {vacinas[i]} : '))
            estoque_por_vacina[i] += qtd
            print(f'O novo estoque de {vacinas[i]} é de {estoque_por_vacina[i]} doses!')

def total_vacinados():
    if len(vacinados.keys()) == 0:
        print('Não há vacinados até o momento.')
    else:
        print(f'Há {len(vacinados.keys())} pessoas vacinadas até o momento, sendo {vacinados_por_vacina[0]} pessoas com Pfizer, {vacinados_por_vacina[1]} pessoas com CoronaVac e {vacinados_por_vacina[2]} pessoas com Astrazeneca.')

def media_diaria():
    if len(vacinados.keys()) == 0:
        print('Não há vacinados até o momento.')
    else:
        print(f'Considerando os últimos 30 dias, temos uma média de {len(vacinados.keys())/30 :.2f} vacinas por dia! Por vacina, temos uma média de {vacinados_por_vacina[0] / 30 :.2f} pessoas com Pfizer, {vacinados_por_vacina[1] / 30 :.2f} pessoas com CoronaVac e {vacinados_por_vacina[2] / 30} pessoas com Astrazeneca.')

def estoque_marca():
    print(f'''Estoque por marca:
Pfizer, estoque: {estoque_por_vacina[0]}
Coronavac, estoque: {estoque_por_vacina[1]}
Astrazeneca, estoque: {estoque_por_vacina[2]}
''')

vacinados = {}
vacinas = ["Pfizer", "CoronaVac", "Astrazeneca"]
vacinados_por_vacina = [0, 0, 0] #pfizer, coronavac, astrazeneca
estoque_por_vacina = [0, 0, 0] #pfizer, coronavac, astrazeneca

ativo = 1
while ativo == 1:
    imprime_menu()

    opcao = int(input('Digite a opção desejada: '))
    while opcao < 1 or opcao > 6:
        print('Opção Inválida')
        opcao = int(input('Digite a opção desejada: '))

    if opcao == 1:
        registrar_vacinacao()
    elif opcao == 2:
        incluir_estoque()
    elif opcao == 3:
        total_vacinados()
    elif opcao == 4:
        media_diaria()
    elif opcao == 5:
        estoque_marca()
    elif opcao == 6:
        print('--- Encerrando programa... ---')
        ativo = 0
```
