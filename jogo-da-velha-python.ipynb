```python
def imprimir_jogo():
    print(f'''
    {tabuleiro[0]} | {tabuleiro[1]} | {tabuleiro[2]}
    {tabuleiro[3]} | {tabuleiro[4]} | {tabuleiro[5]}
    {tabuleiro[6]} | {tabuleiro[7]} | {tabuleiro[8]}
    ''')

def reset_jogo():
    tabuleiro = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    rodada = 0
    vencedor_definido = False

def verificar_vitoria():
    vitoria = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 4, 8], [2, 4,6], [0, 3, 6], [1, 4, 7], [2, 5, 8]]

    for condicao in vitoria: 
        if tabuleiro[condicao[0]] == tabuleiro[condicao[1]] == tabuleiro[condicao[2]]:
            if tabuleiro[condicao[0]] == jogadores[0][1]:
              print(f'PARABÉNS, {jogadores[0][0].upper()} É O VENCEDOR!\n')
              return True
            elif tabuleiro[condicao[0]] == jogadores[1][1]:
              print(f'PARABÉNS, {jogadores[1][0].upper()} É O VENCEDOR!\n')
              return True
              
    return False

def jogar_novamente():
    resposta = -1
    while resposta != "1" and resposta != "0":
        resposta = input("Você deseja jogar novamente? 1 para SIM, 0 para NÃO ")

    return int(resposta)

def jogada():
    if rodada % 2 == 0:
        jogador_atual = jogadores[0]
    else:
        jogador_atual = jogadores[1]
    
    posicao_escolhida = -1
    #impede escolher posição já preenchida ou numero fora do range
    while posicao_escolhida > 9 or posicao_escolhida < 1 or (tabuleiro[posicao_escolhida - 1] == jogadores[0][1] or tabuleiro[posicao_escolhida - 1] == jogadores[1][1]):
        posicao_escolhida = int(input(f'{jogador_atual[0]}, onde deseja marcar com {jogador_atual[1]}? '))

    tabuleiro[posicao_escolhida-1] = jogador_atual[1]

jogo_ativo = 1

while jogo_ativo == 1:
    rodada = 0
    vencedor_definido = False
    jogadores = [["Jogador 1", "●"], ["Jogador 2", "✘"]]
    tabuleiro = [1, 2, 3, 4, 5, 6, 7, 8, 9]

    imprimir_jogo()

    #rodadas
    while rodada < 9 and vencedor_definido == False:
        jogada()
        imprimir_jogo()
        vencedor_definido = verificar_vitoria()
        rodada += 1

    if rodada == 9:
      print("Ihhh, deu VELHA! :P ")
    
    # pergunta se quer jogar novamente
    jogo_ativo = jogar_novamente()
    reset_jogo()

print('----JOGO FINALIZADO----')
```
